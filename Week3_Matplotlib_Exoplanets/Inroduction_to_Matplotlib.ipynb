## Baasic 2D Plots

Upto this point, we have only seen plot of 1D functions ,i.e., 1 input and a single scalar output.

But What about **2D functions**?, i.e., 2-input and 1 scalar output?

WE can use 3D plots, but there are other ways. Like, **Colour Mapping**.. Let's see that

Suppose, we want to plot the function $$z = x^2 + y^2$$

In this case for each value of $x$ and $y$, we get a value of $z$. This is plotted by assiging colours to each pixel or point in the x-y plane.

`messgrid` is used for this. It takes two arrayes eg: [1,2,3] and [6,7,8].

Then it helps to creates each possible ordered pair. Like, [1,6],[1,7],[1,8],[2,6],[2,7]....[3,7],[3,8].

Let's see can example.

```python
x = np.linspace(1,100,10_000)
y = np.linspace(1,100,10_000)
x_v,y_v = np.meshgrid(x,y)
```

```python
x_v
```

```python
y_v
```

If we take x_v and y_v pairwise it will create all coordinate.

```python
z = x_v**2 + y_v**2
```

```python
z
```

```python
plt.contourf(x_v,y_v,z)
plt.show()
```

WE can controll the level and can also see which colour is assigned t

```python
plt.contourf(x_v,y_v,z, levels=50, cmap='plasma')
plt.colorbar(label='Temperature [$^\circ C$]')
plt.xlabel('Horizontal Position [m]')
plt.ylabel('Vertical Position [m]')
```

We can also only draw contours

```python
cs = plt.contour(x_v,y_v,z, levels=20)
plt.clabel(cs, fontsize=8)
```

### Error bars

Sometimes in data we have some error. So, to correctly specify the value we add the error bars.

```python
fig = plt.figure()
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10)

plt.errorbar(x, y + 3, yerr=yerr, label='both limits (default)')

plt.errorbar(x, y + 2, yerr=yerr, uplims=True, label='uplims=True')

plt.errorbar(x, y + 1, yerr=yerr, uplims=True, lolims=True,
             label='uplims=True, lolims=True')

upperlimits = [True, False] * 5
lowerlimits = [False, True] * 5
plt.errorbar(x, y, yerr=yerr, uplims=upperlimits, lolims=lowerlimits,
             label='subsets of uplims and lolims')

plt.legend(loc='lower right',fontsize=12)
plt.show()
```

## Application in Astronomy

Like said before, matplot is used in astronomy for data visualization.

Let's see few examples

```python
from astropy.io import fits
from astropy.wcs import WCS

hdul = fits.open('s0002.fits')
data = hdul[0].data
h1 = hdul[0].header
obj_name = h1.get('OBJECT', 'Unknown')

flux = data[0]
w = WCS(h1, naxis=1, relax=False, fix=False)
wvl = w.wcs_pix2world(np.arange(len(flux)), 0)[0]

plt.plot(wvl, flux)

plt.axvline(x=5895.6, color='red', linestyle='--', label='Na Absorption')
plt.axvline(x=4364.436, color='black', linestyle='--', label=r'$O_{III}$ Emissive')
plt.legend()
plt.xlabel('Wavelength (Angstrom)')
plt.ylabel('Normalized flux')
plt.show()
```

This at once give us so much information. We can see the dips, they are absorption lines. See, just by looking into the data we cannot say much. But now just by looking into this we can say so much about it.

This is not all, we can do more. We can plot H-R diagrams and visualize the data.

```python
from astroquery.gaia import Gaia
from astropy.table import Table
```

```python
tables = Gaia.load_tables(only_names = True)
for table in tables:
    if 'gaiadr3' in table.get_qualified_name():
        print(table.get_qualified_name())
```

```python
query = """SELECT source_id, ra, dec, parallax, phot_g_mean_mag, phot_bp_mean_mag, phot_rp_mean_mag FROM gaiadr3.gaia_source WHERE parallax > 50"""
job = Gaia.launch_job_async(query)
gtable = job.get_results()
```

```python
print(gtable.info)
```

```python
bp_rp = gtable['phot_bp_mean_mag'] - gtable['phot_rp_mean_mag']
fig, ax = plt.subplots(1,1,figsize=(7,7))
ax.hexbin(bp_rp, gtable['phot_g_mean_mag'],bins='log', mincnt=1)
ax.set_xlabel(r"$G_{BP} - G_{RP}$")
ax.set_ylabel(r"G")
ax.set_title("Hertzsprung Russell Diagram Before Filtering from Gaia3")
ax.invert_yaxis()
plt.show()
```

```python
abs_gmag = 5*np.log10(gtable['parallax']) - 10 + gtable['phot_g_mean_mag']
fig, ax = plt.subplots(1,1,figsize=(7,7))
ax.hexbin(bp_rp, abs_gmag,bins='log', mincnt=1)
ax.set_xlabel(r"$G_{BP} - G_{RP}$")
ax.set_ylabel(r"G")
ax.set_title("Hertzsprung Russell Diagram After Filtering from Gaia3")
ax.invert_yaxis()
plt.show()
```

## Exercise 2

The [central limit theorem](http://en.wikipedia.org/wiki/Central_limit_theorem) states that the arithmetic mean of a large 
number of independent random samples (from any distribution) will approach a normal distribution. You can easily test this 
with Numpy and Matplotlib:

1. Create an empty array ``total`` with 10000 values (set to 0)
2. Generate 10000 random values uniformly between 0 and 1
3. Add these values to the ``total`` array
4. Repeat steps 2 and 3 10 times
5. Divide ``total`` by 10 to get the mean of the values you added
5. Make a histogram of the values in ``total``

You can also see how the histogram of ``total`` values changes at each step, if you want to see the evolution!

```python
# your solution here
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create an empty array 'total' with 10000 values set to 0
total = np.zeros(10000)

# Step 2-5: Repeat the process 10 times
for _ in range(10):
    # Step 2: Generate 10000 random values uniformly between 0 and 1
    random_values = np.random.uniform(0, 1, 10000)
    
    # Step 3: Add these values to the 'total' array
    total += random_values

# Step 5: Divide 'total' by 10 to get the mean of the values added
total = total / 10

# Step 6: Make a histogram of the values in 'total'
plt.hist(total, bins=50, edgecolor='black')
plt.title('Histogram of mean values (Central Limit Theorem)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
